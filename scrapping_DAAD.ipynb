# Scrapping DAAD

In this notebook we are going to be able to scrape any international programm from DAAD webpage. We are able to scrape the importante information of each course available in the DAAD webpage.

This Notebook has two main parts:
- Code for get all the information of one programm
- Code to scrappe all the international courses in DAAD

# Code for get all the information of one course

```python
# Clean Code

# Import libraries

from bs4 import BeautifulSoup as bs
import requests

# You need to pass the course_id, you can find this in DAAD web page for example:
# https://www2.daad.de/deutschland/studienangebote/international-programmes/en/detail/4993/ (the ID is 4993)

def get_info_course(course_id):
    # ID
    course_id = course_id

    # Creat dictionary with the info requiered
    search_url = 'https://www2.daad.de/deutschland/studienangebote/international-programmes/en/detail/'+str(course_id)+'/'

    r = requests.get(search_url)

    soup = bs(r.content)


    # Program
    program = soup.select('h2.c-detail-header__title')[0].get_text(' ', strip = True)
    # University
    university = soup.select('h3.c-detail-header__subtitle')[0].get_text(' ', strip = True)

    keys = [list_element.get_text(' ', strip = True).replace('\n               ','') for list_element in soup.select('dt.c-description-list__content')]
    key_values = [list_results.get_text(' ', strip = True)for list_results in soup.select('dd.c-description-list__content.mb-0')]
    
    info_degree = {}
    # Corrrect to Course ID
    
    info_degree['Couse ID'] = course_id
    # Correct to Programme
    info_degree['Program'] = program
    info_degree['University'] = university


    for index, value in enumerate(keys):
        info_degree[value] = key_values[index]
    
    return info_degree
```

```python
# Save data
import json
def save_data(title, data):
    with open(title, 'w', encoding='utf-8') as f:
        json.dump(data, f, ensure_ascii=False, indent=2)
```

```python
# Load Data
import json
def load_data(title):
    with open(title, encoding='utf-8') as f:
        return json.load(f)
```
